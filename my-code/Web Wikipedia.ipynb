```python
data = pd.read_csv('Spotify_v2-1.csv')
```

```python
data.head()
```

```python
artists = list(data['name'])

artists[2] = 'El Cuarteto de Nos'
artists[13] = 'Deadmau5'
artists[30] = 'Ozuna_(singer)'
artists[38] = 'Dimitri_Vegas_&_Like_Mike'
artists[43] = 'Anne-Marie_(singer)'
data['country'] = ''
data['active'] = ''
```

```python
req_titles = ['Origin', 'Years active', 'Years', 'Website']
scrap_col = {'country':['origin'], 'active':['Years active', 'Years'], 'website':['Website']}
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'}
```

```python
for x in artists:
    art = str(x).replace(' ', '_')
    url = 'https://en.wikipedia.org/wiki/' + art
    check_url = requests.get(url)
    status = check_url.status_code
    #time.sleep(1)
    print(status, url)
    #print(check_url, x , url)
    if 200 == status:
        html = check_url.content
        soup = BeautifulSoup(html, 'lxml')
        try:
            table = soup.find_all('table', {'class':'infobox'})[0]
            rows = table.find_all('tr')
            for y in rows:
                if 'Origin' in str(y):
                    country = y.td.text
                    data['country'][data.name == x] = country
                    #print(check_url, x, ' - Pais:', y.td.text)   
                    continue
        except:
            print(':::::::')
            continue
        try:
            table = soup.find_all('table', {'class':'infobox'})[0]
            rows = table.find_all('tr')
            for y in rows:
                if 'Born' in str(y):
                    country = y.td.a.text
                    data['country'][data.name == x] = country
                    #print(check_url, x, ' - Pais:', y.td.a.text)   
                    continue
        except:
            print(':::::::')
            continue
        try:
            html = check_url.content
            soup = BeautifulSoup(html, 'lxml')
            table = soup.find_all('div', {'class':'birthplace'})[0]
            country = table.a.text
            data['country'][data.name == x] = country
            continue
        except:
            print(':::::::')
            continue
        
    else:
        pass
```
