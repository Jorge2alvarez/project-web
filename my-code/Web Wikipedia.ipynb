```python
for x in artists:
    art = str(x).replace(' ', '_')
    url = 'https://en.wikipedia.org/wiki/' + art
    check_url = requests.get(url)
    status = check_url.status_code
    #time.sleep(1)
    #print(status, url)
    #print(check_url, x , url)
    if 200 == status:
        html = check_url.content
        soup = BeautifulSoup(html, 'lxml')
        try:
            table = soup.find_all('table', {'class':'infobox'})[0]
            rows = table.find_all('tr')
            for y in rows:
                if 'Origin' in str(y):
                    print(check_url, x, url, '\nPais:', y.td.text)   
                    pass
        except:
            print(':::::::')
            continue
        try:
            html = check_url.content
            soup = BeautifulSoup(html, 'lxml')
            table = soup.find_all('div', {'class':'birthplace'})[0]
            print(check_url, x, url, '\nPais:', table.text)     
        except:
            print(':::::::')
            continue
        
    else:
        print(check_url, x, "\nError al cargar::::::::::::::::::")
```
